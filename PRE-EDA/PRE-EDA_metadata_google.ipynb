```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter
```

```python
#Funcion para imprimir un ejemplo no nulo de cada variable
def imprimir_ejemplos(df):
    ejemplo = []
    titulo = []
    tipo = []
    global titulos
    global tipos
    for elemento in df.columns:
        for valor in df[elemento]:
            flag = 0
            while valor != None:
                ejemplo.append(valor)
                titulo.append(elemento)
                tipo.append(type(valor))
                flag = 1
                break
            if flag == 1:
                break
    titulos.append(titulo)
    tipos.append(tipo)
    for i in range(len(ejemplo)):
        print(titulo[i], ': ', ejemplo[i])
        print('Tipo de dato: ', tipo[i])
        print()
```

```python
#Nombre de la carpeta donde estan alojados los archivos
carpeta = 'C:/Users/eduen/Desktop/PF_Yelp-Maps_DS-04pt/metadata_google'

#En lista_df se carga cada dataframe para luego concatenarlos
lista_df = []

#Titulos y tipos cargan por cada dataframe una lista con los nombres de las columnas y el tipo de dato de cada una
#Se van a Utilizar para comparar que sean equivalentes en todos los df tanto los nombres como los tipos de dato
titulos = []
tipos = []

#Este for itera por los nombres de los archivos dentro de la carpeta
for archivo in os.listdir(carpeta):
   
    df = pd.read_json(f'{carpeta}/{archivo}', lines=True)#Carga el df
    lista_df.append(df)#Lo guarda en la lista para concatenar

    print(f'ANALISIS PRELIMINAR DATASET: {archivo}')
    print()
    print('EJEMPLOS NO NULOS:')
    print()
    display(imprimir_ejemplos(df))#Imprime el primer dato NO NULO de cada columna y el tipo de dato del mismo
```

```python
idem_titulos = 'Las columnas son equivalentes' #Variable simil a un flag indicando que en todos los df las columnas son identicas en nombres y cantidad
idem_tipos = 'Los tipos de datos son equivalentes' #Idem anterior
#Se itera la lista titulos, se da por hecho que la lista tipos tiene la misma longitud ya que surgen de la iteracion de dataframes
for i in range(1, len(titulos)):#Iteracion arrancando por el segundo elemento de la lista, se comparara con el elemento anterior
    if titulos[i] != titulos[i - 1]:
        idem_titulos = 'Hay diferencias' #Si encuentra diferencia cambia el flag de titulos

    if tipos[i] != tipos[i - 1]:
        idem_tipos = 'Hay diferencias' #Si encuentra diferencia cambia el flag de tipos

print('Titulos: ', idem_titulos)
print()
print('Tipos de datos: ', idem_tipos)
```

```python
df_final = pd.concat(lista_df, ignore_index=True)#Concatenacion de todos los df
df_final.to_parquet('metadata_google.parquet')#Cargar en un archivo .parquet
```

```python
lista_categorias = []
def extraer_category(dato):
    if type(dato) == list:
        for elemento in dato:
            lista_categorias.append(elemento)
    
    return dato

df_uni['category'].apply(extraer_category)

print(len(lista_categorias))
```

```python
categorias = Counter(lista_categorias)
wordcloud = WordCloud(width=1000, height=600, background_color='white').generate_from_frequencies(categorias)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.tight_layout()
plt.show()
```

Se grafico una nube de palabras basado en la frecuencia de repeticiones de las diferentes categorias, arrojando visualmente el grafico que la categoria Restaurant es la que clasifica mayor numero de establecimiento entre las mismas.
Campo analizado: category

```python
top_tres = categorias.most_common(3)
print(f'Total de establecimientos con categoria {top_tres[0][0]}: {top_tres[0][1]}')
print(f'Total de establecimientos con categoria {top_tres[1][0]}: {top_tres[1][1]}')
print(f'Total de establecimientos con categoria {top_tres[2][0]}: {top_tres[2][1]}')
```

category: Se puede observar que existen 97257 establecimientos encuadrados en la categoria Restaurant, 77854 en la categoria Auto Repair Shop y $70143 en la categoria Gas Station, siendo Restaurant la categoria de la cual existe mayor cantidad de establecimientos.
