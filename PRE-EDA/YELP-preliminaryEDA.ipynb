<b>DATA TIPS</b>

```python
df2 = pd.read_json('tip.json', lines=True)
pd.DataFrame(df2)
```

```python
df2.info()
```

```python
df2['compliment_count'].value_counts()
```

```python
df2['date'] = pd.to_datetime(df2['date'])
df2['year'] = df2['date'].dt.year

reseñas_por_año = df2.groupby('year').size()

plt.figure(figsize=(12, 6))
reseñas_por_año.plot(kind='bar', color='skyblue')
plt.title('Cantidad de Reseñas por Año')
plt.xlabel('Año')
plt.ylabel('Cantidad de Reseñas')
plt.show()
```

Este gráfico nos muestra la cantidad de sugerencias que se realizaron a los comercios, podemos ver como fue la interacción de los usuarios a lo largo de los  años y pensar en que, quizás, la disminucion de sugerencias, indica una mejora en varios aspectos los comercios. 

Este dataset nos brinda los consejos que los usuarios hicieron a los negocios, con la fecha en formato datetime, el comentario en si y un "recuento de cumplidos". El archivo está bastante bien por lo que se ve (no hay nulos), y se puede relacionar con business.plk mediante el business_id. Se puede aplicar NLP para ver cuáles son los comercios con sugerencias negativas/positivas/neutrales y agregarle esa clasificación para conocer sus deficiencias.

```python
df_filtrado = df4[df4['fans'] <= 4000]

plt.figure(figsize=(10, 6))
plt.scatter(df_filtrado['fans'], df_filtrado['review_count'], alpha=0.5, color='blue')

plt.title('Relación entre Fans y Review Count (sin outliers)')
plt.xlabel('Fans')
plt.ylabel('Review Count')

plt.show()
```

En este gráfico se busca evaluar una relación entre cantidad de fans y cantidad de reviews de un usuario, con el fin de conocer su alcance en la plataforma.

Este dataset nos proporciona información sobre los usuarios de Yelp, la cantidad de reseñas que hicieron y el valor de ellas, si son usuarios "elite" (es decir, que tienen varias interacciones), nos puede indicar el alcance que tienen sus reacciones por la cantidad de fans o seguidores que tiene, desde cuándo es usuario. También podemos conocer usuarios relacionados (friends) y cómo reacciona ante otras reseñas.

<b>DATA REVIEWS</b>

```python
chunk_size = 1000000

df5 = pd.DataFrame()

chunks = pd.read_json('review.json', lines=True, chunksize=chunk_size)
for chunk in chunks:

    resultados = pd.concat([df5, chunk], ignore_index=True)

    break
```

```python
resultados
```

```python
resultados.info()
```

```python
columns = ['useful', 'funny', 'cool']
recuento = resultados[columns].sum()
recuento
```

```python
bars = plt.bar(columns, recuento, color=['blue', 'orange', 'green'])

plt.xlabel('Reacciones')
plt.ylabel('Recuento')
plt.title('Recuento por Reacciones')

plt.ticklabel_format(style='plain', axis='y')
plt.bar_label(bars, fmt='%d', fontsize=9, label_type='edge', color='black')

plt.show()
```

En este gráfico podemos ver cuál es la reacción más utilizada dentro de la muestra de nuestro dataset. 

En este dataset podemos ver las reseñas que realizaron los usuarios (es util para hacer el analisis de sentimientos) junto con la cantidad de estrellas, la fecha y el usuario que la hizo. Debido al tamaño del archivo, se fragmentó para este primer análisis en el primer millón de filas. En este caso, no se hallaron datos nulos.
