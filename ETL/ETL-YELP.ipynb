<b>DATA CHECK IN</b>

```python
df8 = pd.read_json('checkin.json', lines=True)
pd.DataFrame(df8)
```

<p>Se desarma la columna con las fechas.</p>

```python
df8['date'] = df8['date'].apply(lambda x: x.split(','))
df_explode = df8.explode('date')
df_explode['date'] = pd.to_datetime(df_explode['date'].str.strip(), format='%Y-%m-%d %H:%M:%S', errors='coerce')
df_explode
```

<p>Filtra por los años de muestra.</p>

```python
df8 = df_explode[(df_explode['date'].dt.year >= 2018) & (df_explode['date'].dt.year <= 2021)]
```
